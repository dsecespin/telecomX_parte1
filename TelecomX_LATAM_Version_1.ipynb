```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
df = pd.read_json(url)
df
```

```python
 cliente = pd.json_normalize(df['customer'])
 cliente
```

```python
fono = pd.json_normalize(df['phone'])
fono
```

```python
internet = pd.json_normalize(df['internet'])
internet
```

```python
cuenta = pd.json_normalize(df['account'])
cuenta
```

```python
columnas = list(df.columns)
columnas
```

```python
df_dropped = df.drop(['customer','phone','internet','account'],axis=1)
df_dropped
dato = pd.concat([df_dropped,cliente,fono,internet,cuenta],axis=1)
dato
```
